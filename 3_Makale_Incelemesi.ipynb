```python
import matplotlib.pyplot as plt
import seaborn as sns

# Veriler
saldiri_turleri = ["Kaba Kuvvet", "DDoS", "DoS", "MITM", "Tekrar"]
algilama_oranlari = [98, 96, 98, 96, 96]

# Çubuk grafik
plt.figure(figsize=(10, 6))
sns.barplot(x=saldiri_turleri, y=algilama_oranlari, palette="viridis")
plt.title("Deep-IDS Algılama Oranı (Sızma Türüne Göre)")
plt.xlabel("Sızma Türü")
plt.ylabel("Algılama Oranı (%)")
plt.ylim(0, 100)

# Grafiği kaydet
plt.savefig('deep_ids_algilama_orani.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
labels = ['Veri Güncelliği', 'Pratiklik Eksikliği', 'Vaka Çalışması Eksikliği']
deep_ids = [8, 3, 5]
blockchain = [4, 7, 6]
zta = [3, 6, 8]

x = range(len(labels))
width = 0.25

plt.bar([p - width for p in x], deep_ids, width=width, label='Deep-IDS')
plt.bar(x, blockchain, width=width, label='Blockchain')
plt.bar([p + width for p in x], zta, width=width, label='ZTA')

plt.xticks(x, labels)
plt.title("Zayıf Yönlerin Karşılaştırması (1: Az, 10: Çok)")
plt.ylabel("Etkilenme Düzeyi")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veriler
kategoriler = ["Kimlik/Erişim", "IoT Güvenliği", "Veri Bütünlüğü", "Sızma Algılama", "Diğer"]
makale_sayilari = [30, 25, 15, 10, 31]
renkler = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]

# Pasta grafik
plt.figure(figsize=(8, 8))
plt.pie(makale_sayilari, labels=kategoriler, colors=renkler, autopct='%1.1f%%', startangle=140)
plt.title("Blockchain Kullanım Senaryoları Dağılımı (111 Makale)")

# Grafiği kaydet
plt.savefig('blockchain_kullanim_senaryolari.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Veriler (örnek metrikler)
makaleler = ["Deep-IDS", "ZTA İncelemesi", "Blockchain SLR"]
metrikler = {
    "Doğruluk/Kapsam": [97.67, 80, 90],  # Deep-IDS: % doğruluk, ZTA: varsayımsal kapsam, Blockchain: varsayımsal kapsam
    "Pratik Uygulama": [90, 50, 60]      # Varsayımsal pratiklik puanı (100 üzerinden)
}

# DataFrame oluştur
df = pd.DataFrame(metrikler, index=makaleler)

# Çubuk grafik
plt.figure(figsize=(10, 6))
df.plot(kind='bar', color=['#1f77b4', '#ff7f0e'], ax=plt.gca())
plt.title("Makalelerin IoT Güvenliğine Katkı Karşılaştırması")
plt.xlabel("Makaleler")
plt.ylabel("Puan (0-100)")
plt.legend(title="Metrik")
plt.xticks(rotation=45)
plt.tight_layout()

# Grafiği kaydet
plt.savefig('makale_karsilastirma.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Yanıt süreleri
saldiri_turleri = ["Kaba Kuvvet", "DDoS", "DoS", "MITM", "Tekrar"]
yanit_sureleri = [1.46, 2.02, 1.98, 1.05, 0.92]

# Kutu grafik
plt.figure(figsize=(10, 6))
sns.boxplot(x=saldiri_turleri, y=yanit_sureleri, palette="Set2")
plt.title("Deep-IDS Yanıt Süreleri (Sızma Türüne Göre)")
plt.xlabel("Sızma Türü")
plt.ylabel("Yanıt Süresi (saniye)")

# Grafiği kaydet
plt.savefig('deep_ids_yanit_sureleri.png', dpi=300, bbox_inches='tight')
plt.show()
```
